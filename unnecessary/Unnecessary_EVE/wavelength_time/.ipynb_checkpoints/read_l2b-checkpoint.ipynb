```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a sample dataframe
data = np.random.rand(10, 5)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# Visualize the dataframe using imshow
fig, ax = plt.subplots()
im = ax.imshow(df.values, cmap='YlOrBr')

# Add colorbar to the plot
cbar = ax.figure.colorbar(im, ax=ax)

# Set tick labels
ax.set_xticks(np.arange(len(df.columns)))
ax.set_yticks(np.arange(len(df.index)))
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.index)

# Rotate the tick labels and set their alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations
for i in range(len(df.index)):
    for j in range(len(df.columns)):
        text = ax.text(j, i, round(df.values[i, j], 2),
                       ha="center", va="center", color="black")

# Set title
ax.set_title("Sample Dataframe")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# create the y tick values array
y_tick_values = [3.000e+01, 6.030e+03, 1.203e+04, 1.803e+04, 2.403e+04, 3.003e+04,
                 3.603e+04, 4.203e+04, 4.803e+04, 5.403e+04, 6.003e+04, 6.603e+04,
                 7.203e+04, 7.803e+04, 8.403e+04]

# create the plot and set the y tick values and format
fig, ax = plt.subplots()
ax.plot([0,1])
ax.set_yticks(y_tick_values)
ax.set_yticklabels(['{:.2e}'.format(val) for val in y_tick_values])
plt.show()
```
