```python
import astropy.units as u
from sunpy.net import Fido, attrs as a
import sunpy.map
```

## query for one month

```python
obs_start = "2011/01/28 00:00:00"
obs_end = "2011/02/28 00:00:00"
obs_time = a.Time(obs_start, obs_end)

results = Fido.search(obs_time,
                      a.Instrument.hmi,
                      a.Sample(24*60*u.min),
                     ) # hmi
```
