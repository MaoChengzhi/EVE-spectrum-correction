## example of Nealder Mead

```python
def f(x):
    a,b,c=x
    return (a**2+b**2+c**2)

x0=[2,3,4]

num_evals=0

def callback(x):
    global num_evals
    num_evals += 1
    print(f"Iteration {num_evals}: error_function({x}) = {f(x)}")


    
result = minimize(f, x0,     callback=callback                 ,
                  method='Nelder-Mead')
```

```python
def f(x):
    a,b,c=x
    return (a**2+math.sin(b)**2+c**2+1)*(a-1)**2

x0=[2,3,4]

num_evals=0

def callback(x):
    global num_evals
    num_evals += 1
    print(f"Iteration {num_evals}: error_function({x}) = {f(x)}")

options = {'adaptive': True,"return_all":True}
    
result = minimize(f, x0,     callback=callback                 ,
                  method='Nelder-Mead',options=options)
```
