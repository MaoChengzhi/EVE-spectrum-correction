visualization of wavelength consistent simualtion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi
import math
from astropy.constants import au, R_sun
```

```python
transunit = ((au/R_sun).value)**2 / 1000
angle_point_num_alpha = 61
offaxis_angle_x_alpha = np.linspace(-math.pi /
                                    360, math.pi/360, angle_point_num_alpha)
offaxis_angle_y_alpha = np.zeros(angle_point_num_alpha)
```

```python
DN=np.load("discrete_DN.npz")
```

```python
DN["solution"][:,0]
```

```python
fig, ax = plt.subplots()
ax.plot(offaxis_angle_x_alpha*180*60/pi,DN["solution"][:,0],label="Simulation")
ax.plot(offaxis_angle_x_alpha*180*60/pi,19.8*transunit*offaxis_angle_x_alpha**2,
        label="Chamberlin obs")
ax.set_xlim(-30,30)
ax.set_title("Unpanned simulation")

ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(offaxis_angle_x_alpha*180*60/pi,DN["solution"][:,0]-DN["solution"][:,0].min(),label="Simulation")
ax.plot(offaxis_angle_x_alpha*180*60/pi,19.8*transunit*offaxis_angle_x_alpha**2,
        label="Chamberlin obs")
ax.set_xlim(-30,30)

ax.set_ylabel('Wavelength shift (nm)')
ax.set_xlabel("Alpha Offset (arcmin)")
ax.set_title("Panned simulation")

ax.legend()
```
