## 将波长和 像素 都连续化的结果
 二维积分取到4阶近似

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

## degree 4

```python
df = pd.read_csv("myfile_degree4.csv", header=None)
alpha=df.iloc[0]
```

```python
df.iloc[0]
```

```python
type(df.iloc[1][0])
```

```python
fig, ax = plt.subplots()
ax.plot(df.iloc[0],df.iloc[1])
```

```python
fig, ax = plt.subplots()
ax.scatter(df.iloc[0]*180*60/pi,df.iloc[1])
ax.set_xlim(-30,30)
```

```python
fig, ax = plt.subplots()
ax.plot(df.iloc[0]*180*60/pi,df.iloc[1],label="output simulation")
ax.plot(df.iloc[0]*180*60/pi,915.52*df.iloc[0]**2,label="input A=19.8")
ax.set_xlim(-30,30)
ax.set_ylabel('Wavelength shift (nm)')
ax.set_xlabel("Alpha Offset (arcmin)")
ax.set_title("Upanned simulation")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(df.iloc[0]*180*60/pi,df.iloc[1]-df.iloc[1].min(),label="output simulation")
ax.plot(df.iloc[0]*180*60/pi,915.52*df.iloc[0]**2,label="input A=19.8")
ax.set_xlim(-30,30)
ax.set_ylabel('Wavelength shift (nm)')
ax.set_xlabel("Alpha Offset (arcmin)")
ax.set_title("Panned simulation")
ax.legend()
```
