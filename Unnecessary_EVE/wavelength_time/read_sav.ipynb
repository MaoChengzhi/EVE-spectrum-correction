```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'Country': ['USA', 'Canada', 'Mexico', 'Japan', 'China', 'India'],
        'Population': [327.2, 37.6, 130.2, 126.5, 1392.7, 1369.0],
        'GDP': [21.4, 1.8, 1.1, 4.9, 14.1, 2.6]}

df = pd.DataFrame(data)

# Create a bar chart of population by country
fig, ax = plt.subplots()
df.plot(x='Country', y='Population', kind='bar', ax=ax)

# df.plot(x='Country', y='GDP', kind='bar', ax=ax)

# Set the chart title and axis labels
ax.set_title('Population by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Population (millions)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'Country': ['USA', 'Canada', 'Mexico', 'Japan', 'China', 'India'],
        'Population': [327.2, 37.6, 130.2, 126.5, 1392.7, 1369.0],
        'GDP': [21.4, 1.8, 1.1, 4.9, 14.1, 2.6]}

df = pd.DataFrame(data)

# Create a bar chart of population by country
fig, ax = plt.subplots()
# df.plot(x='Country', y='Population', kind='bar', ax=ax)

df.plot(x='Country', y='GDP', kind='bar', ax=ax)

# Set the chart title and axis labels
ax.set_title('GDP by Country')
ax.set_xlabel('Country')
ax.set_ylabel('GDP')

plt.show()
```
