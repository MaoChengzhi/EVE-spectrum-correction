## SSIM

```python
def calculate_ssim(image1, image2):
    # Ensure that the images are in the range [0, 255] and are of dtype 'uint8'
    image1 = np.clip(image1, 0, 255).astype(np.uint8)
    image2 = np.clip(image2, 0, 255).astype(np.uint8)

    # Calculate SSIM between the two images
    ssim_score, _ = ssim(image1, image2, full=True)

    return ssim_score
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample images (replace these with your actual image arrays)
image_array1 = np.random.rand(256, 256)  # Random 256x256 image
image_array2 = np.random.rand(256, 256)  # Another random 256x256 image
# image_array2=image_array1+90.9*np.random.rand(256, 256)



# Calculate SSIM
ssim_score = calculate_ssim(image_array1, image_array2)
print(f"SSIM Score: {ssim_score}")

# Display the images (optional)
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image_array1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_array2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim

# Sample images (replace these with your actual image arrays)
image_array1 = np.random.rand(256, 256) * 255  # Random 256x256 image
image_array2 = np.random.rand(256, 256) * 255  # Another random 256x256 image

def calculate_ssim(image1, image2):
    # Ensure that the images are in the range [0, 255] and are of dtype 'uint8'
    image1 = np.clip(image1, 0, 255).astype(np.uint8)
    image2 = np.clip(image2, 0, 255).astype(np.uint8)

    # Calculate SSIM between the two images
    ssim_score, _ = ssim(image1, image2, full=True)

    return ssim_score

# Calculate SSIM
ssim_score = calculate_ssim(image_array1, image_array2)
print(f"SSIM Score: {ssim_score}")

# Display the images (optional)
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image_array1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_array2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.show()
```
